```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot")

x=[1,2,4,6]
y=[7,8,10,12]
x2 =[8,9,10,11]
y2 =[10,12,14,16]
plt.plot(x,y, label='first',linewidth =3)
plt.plot(x2,y2,label = 'second',linewidth = 3)
plt.title("hey world")
plt.xlabel("this is x-axis")
plt.ylabel("this is y-axix")
plt.legend() #it helps to denote the lines into specific word i.e red,blue comes out first and second
plt.show()
```
