# fillna()

```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Aman\\Downloads\\Fortune_10.csv")
df
```

```python
df.fillna(3)
```

```python
df.fillna({'Name':'Pagal','Profit': 3002282,'Growth': 0})
```

```python
df.fillna(method = 'ffill') #this print the forward value from thr nun fn.
```

```python
df.fillna(method = 'bfill') #this print the backward value from thr nun fn.
```

```python
df.fillna(method = 'pad') #this print the forward value from thr nun fn. same as ffill method
```

```python
df.fillna(method = 'ffill', axis = 1 ) #(for axis fn. depends on the row or column)
```

```python
df.fillna(method = 'ffill', axis = 0 ) #(for axis fn. depends on the row or column)
```

```python
df.fillna(3,limit = 1 )
```

```python
df.fillna(3,limit = 2 ) #same as use for bfill and ffill
```

```python
df.fillna(3,inplace = True) #to update the file
df
```
