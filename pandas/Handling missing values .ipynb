# handling missing value#3

```python
import pandas as pd
```

```python
df2 =  pd.read_csv("C:\\Users\\Aman\\Downloads\\Fortune_10.csv")
df2
```

```python
df2.dropna() #drop every rows which have a nan value (one or multiple)
```

```python
df2.dropna(axis = 1) #axis = 1,this drops all the nan values from the columns.#axis= 0,this drops the values of the funcitons in the rows.
```

```python
df2.dropna(how = 'any') #this is similar to the simple dropna functions, if we can't write axis the it by defaults take axis =0 it means row wise
```

```python
df2.dropna(how = 'all')#this only drop that rows or column which all values is null or nan or axis functions is same as above fn.
```

```python
df2.dropna(thresh =1)#this only print the not null values depend upon the conditions
```

```python
df2.dropna(subset = ['Name'])#this only print those values which are not null
```
